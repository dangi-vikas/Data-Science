**Sigmoid Function**

A *sigmoid function* (Logistic Function) is a bounded, differentiable, real function that is defined for all real input values and has a range of (0,1).

<br>
<div>It is used to calculate the probablity of an oblect belonging to a certain class.</div>
<br>
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x_axis = np.linspace(-10, 10, 100)
Z = 1 / (1 + np.exp(-x_axis))

plt.plot(x_axis, Z)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")

plt.show()
```

**Compute the Cost** $J(\omega)$

The objective of logistic regression is to minimise the cost function 

$$J(\omega) = - \frac{1}{m} \sum_{i=1}^{m} [y_i * log⁡(h_\omega(x_i)) + (1 - y_i) * log⁡(1 - (h_\omega(x_i))]$$

where h(x) is the hypothesis and given by the linear model

 $$h_\omega(x) = \omega_0 + \omega_1 x_1$$

```python
def cost_function(X, y, omega):
    m = len(y)
    y_pred = sigmoid(np.dot(X, omega))
    error = (y * np.log(y_pred)) + ((1 - y) * np.log(1 - y_pred))
    cost = -1 / m * sum(error)
    gradient = 1 / m * np.dot(X.transpose(), (y_pred - y))
    return cost[0], gradient
```

**Standardisation**

Standardisation refers to the process of subtracting the mean and dividing by the standard deviation.

<br>
$$Z = \frac{x - \mu}{\sigma}$$

```python
#Standardisation

mean_scores =  np.mean(scores, axis=0)
std_scores =  np.std(scores, axis=0)

scores = (scores - mean_scores) / std_scores
```

```python
rows = scores.shape[0]
cols = scores.shape[1]

X = np.append(np.ones((rows, 1)), scores, axis=1)
y = results.reshape(rows, 1)

omega = np.zeros((cols + 1, 1))
cost, gradient = cost_function(X, y, omega)
```

**Gradient Descent**


The gradient descent of the cost function is given by

$$\frac{\partial J(\omega)}{\partial \omega} = \frac{1}{m} \sum_{i=1}^{m} (h_{\omega}(x_i) - y_i)x_i$$

Minimize the cost function $J(\omega)$ by updating the below equation and repeating until convergence, simultaneously updating  $\omega_j$ for all j

$$\omega: = \omega - \alpha \frac{\partial J(\omega)}{\partial \omega}$$

```python
def gradient_descent(X, y, omega, alpha, epochs):
    costs = []
    
    for i in range(epochs):
        cost, gradient = cost_function(X, y, omega)
        omega -= (alpha * gradient)
        costs.append(cost)
        
    return omega, costs
```

```python
omega, costs = gradient_descent(X, y, omega, 1, 200)
```

**Plotting the convergence of** *J(ω)*

Plot $J(\theta)$ against the number of iterations of gradient descent:

```python
plt.plot(costs)
plt.xlabel("Iterations")
plt.ylabel("$J(\Theta)$")
plt.title("Values of Cost Function over different iterations of Gradient Descent")
```

**Predicting the result using optimised** $\omega$ **values**

```python
def predict(omega, x):
    results = x.dot(omega)
    return results > 0 
```
